```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Prepare Data

```python
plt.rc('figure', figsize=(10, 5), dpi=120) # Đặt global figsize, dpi trong plt.figure

df = pd.read_csv('dm_office_sales.csv')
df.head()
```

## Distribution Plots - Rug plot

```python
sns.rugplot(data=df, x='salary')
```

```python
sns.rugplot(data=df, x='salary', height=0.5) # height là chiều cao 50% theo trục y
```

## Distribution Plots - Histogram

```python
# Cú pháp: distplot đã deprecated, thay vào đó sử dụng displot
```

```python
sns.histplot(data=df, x='salary')
```

```python
sns.histplot(data=df, x='salary', bins=50) # Có thể thay đổi số lượng bins
```

```python
sns.set(style='darkgrid') # high level command, khi dùng rồi sẽ thay đổi vĩnh viễn cho các biểu đồ sau
sns.histplot(data=df, x='salary', bins=40)
```

```python
# Thay đổi màu nền vĩnh viễn khi dùng sns.set ở trên
sns.histplot(data=df, x='salary', bins=40)
```

```python
sns.set(style='white') # Đặt trở lại background mặc định
sns.histplot(data=df, x='salary', bins=40)
```

```python
sns.histplot(data=df, x='salary', color='red', edgecolor='green', lw=5, ls=':') # Có thể thay đổi color và style
```

```python
sns.displot(data=df, x='salary', height=5, aspect=2) # Nên dùng displot (dùng FaceGrid thay vì Figure nên không bị ảnh hưởng plt.figure)
```

```python
plt.figure(figsize=(10, 5), dpi=120)
sns.displot(data=df, x='salary', height=5, aspect=2, kde=True)
```

```python
sns.displot(data=df, x='salary', height=5, aspect=2, rug=True)
```

## Distribution Plots - KDE Plot

```python
sns.kdeplot(data=df, x='salary')
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0, 100, 200)
sample_ages
```

```python
sample_ages = pd.DataFrame(sample_ages, columns=['age'])
sample_ages.head(10)
```

```python
sns.displot(data=sample_ages, x='age', height=5, aspect=2, bins=30, rug=True, kde=True)
```

```python
sns.kdeplot(data=sample_ages, x='age')
```

```python
# clip=[min, max] sẽ cắt bỏ bất kỳ giá trị nào nằm ngoài khoảng này khi vẽ đường KDE
sns.kdeplot(data=sample_ages, x='age', clip=[0, 100])
```

```python
# bw_adjust (default=1) parameter sử dụng để điều chỉnh độ rộng của băng tần (bandwidth)
sns.kdeplot(data=sample_ages, x='age', clip=[0, 100], bw_adjust=0.1) # Giá trị bw_adjust nhỏ (<1), KDE chi tiết hơn nhưng có thể bị nhiễu
```

```python
sns.kdeplot(data=sample_ages, x='age', clip=[0, 100], bw_adjust=1.25) # Giá trị bw_adjust lớn (>1), KDE mượt hơn nhưng có thể làm mất chi tiết
```

```python
# fill=True được sử dụng để tô màu phần bên dưới đường KDE
sns.kdeplot(data=sample_ages, x='age', clip=[0, 100], bw_adjust=0.1, fill=True) 
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Prepare Data

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

## Scatterplots

```python
plt.figure(figsize=(14, 5), dpi=150) # Seaborn tng thch vi syntax ca Matplotlib

# hue parameter nhm d liu theo mt categorical feature v hin th mu sc khc nhau cho tng nhm
# hue parameter s dng khi mun phn bit d liu da trn mt danh mc nh gii tnh, trnh  hc vn, loi sn phm
# palette parameter c s dng  ch nh bng mu khi s dng hue: https://matplotlib.org/stable/gallery/color/colormap_reference.html
sns.scatterplot(x='salary', y='sales', data=df, hue='level of education', palette='cool')
```

```python
plt.figure(figsize=(14, 5), dpi=150)
sns.scatterplot(x='salary', y='sales', data=df, hue='salary') # Gradient
```

```python
plt.figure(figsize=(14, 5), dpi=150)

# size paramter iu chnh kch thc im da trn mt continuous feature
sns.scatterplot(x='salary', y='sales', data=df, size='work experience')
```

```python
plt.figure(figsize=(14, 5), dpi=150)

# s paramter thit lp kch thc c nh cho tt c cc im trong biu 
sns.scatterplot(x='salary', y='sales', data=df, s=150)
```

```python
plt.figure(figsize=(14, 5), dpi=150)

# alpha parameter iu chnh  trong sut gia cc im ( d nhn khi cc im xp chng ln nhau)
sns.scatterplot(x='salary', y='sales', data=df, s=150, alpha=0.2)
```

```python
plt.figure(figsize=(14, 5), dpi=150)

# style parameter  thay i hnh dng (marker style) ca cc im da trn mt categorical feature
sns.scatterplot(x='salary', y='sales', data=df, s=80, style='level of education')
```

```python
plt.figure(figsize=(14, 5), dpi=150)

# Kt hp gia style v hue
sns.scatterplot(x='salary', y='sales', data=df, s=80, style='level of education', hue='level of education')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Prepare Data

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

## Scatterplots

```python
plt.figure(figsize=(14, 5), dpi=150) # Seaborn tương thích với syntax của Matplotlib

# hue parameter nhóm dữ liệu theo một categorical feature và hiển thị màu sắc khác nhau cho từng nhóm
# hue parameter sử dụng khi muốn phân biệt dữ liệu dựa trên một danh mục như giới tính, trình độ học vấn, loại sản phẩm
# palette parameter được sử dụng để chỉ định bảng màu khi sử dụng hue: https://matplotlib.org/stable/gallery/color/colormap_reference.html
sns.scatterplot(x='salary', y='sales', data=df, hue='level of education', palette='cool')
```

```python
plt.figure(figsize=(14, 5), dpi=150)
sns.scatterplot(x='salary', y='sales', data=df, hue='salary') # Gradient
```

```python
plt.figure(figsize=(14, 5), dpi=150)

# size paramter điều chỉnh kích thước điểm dựa trên một continuous feature
sns.scatterplot(x='salary', y='sales', data=df, size='work experience')
```

```python
plt.figure(figsize=(14, 5), dpi=150)

# s paramter thiết lập kích thước cố định cho tất cả các điểm trong biểu đồ
sns.scatterplot(x='salary', y='sales', data=df, s=150)
```

```python
plt.figure(figsize=(14, 5), dpi=150)

# alpha parameter điều chỉnh độ trong suốt giữa các điểm (để dễ nhìn khi các điểm xếp chồng lên nhau)
sns.scatterplot(x='salary', y='sales', data=df, s=150, alpha=0.2)
```

```python
plt.figure(figsize=(14, 5), dpi=150)

# style parameter để thay đổi hình dạng (marker style) của các điểm dựa trên một categorical feature
sns.scatterplot(x='salary', y='sales', data=df, s=80, style='level of education')
```

```python
plt.figure(figsize=(14, 5), dpi=150)

# Kết hợp giữa style và hue
sns.scatterplot(x='salary', y='sales', data=df, s=80, style='level of education', hue='level of education')
```

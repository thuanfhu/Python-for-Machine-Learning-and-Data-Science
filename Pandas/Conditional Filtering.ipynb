```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('tips.csv')
df.head()
```

```python
df['total_bill'] > 45 # Boolean Series
```

```python
type(df['total_bill'] > 45)
```

```python
df[df['total_bill'] > 45] # Lọc những rows có total_bill > 45
```

```python
# AND &  --- Cả hai điều kiện cần phải đúng
# OR |  --- Chỉ cần một trong hai điều kiện đúng
```

```python
df[(df['total_bill'] > 40) & (df['day'] == "Sun")] # Lọc với nhiều điều kiện, nếu dùng and/or (Python key) sẽ lỗi ambiguous vì mỗi vế là Boolean Series
```

```python
# Thay vì (df['day'] == "Fri") | (df['day'] == "Sat") | (df['day'] == "Sun")
dayOptions = ['Fri', 'Sat', 'Sun']
df[(df['day'].isin(dayOptions)) & (df['total_bill'] > 40)] # Cú pháp ngắn gọn hơn với isin
```

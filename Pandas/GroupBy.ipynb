## GroupBy Operations - Basics

```python
df['model_year'].value_counts() # Đếm số lần xuất hiện của mỗi năm trên col model_year
```

```python
df.groupby('model_year') # Lazy object bởi vì chưa có aggregate function
```

```python
df.groupby('model_year').mean(numeric_only=True) # numeric_only=True, pandas chỉ tính mean trên các cột có kiểu số (int, float)
```

```python
df.groupby(['model_year', 'cylinders']).mean(numeric_only=True).head(12)
```

```python
df.groupby(['model_year', 'cylinders']).mean(numeric_only=True).columns
```

```python
df.groupby(['model_year', 'cylinders']).mean(numeric_only=True).index
```

```python
df.groupby(['model_year', 'cylinders']).mean(numeric_only=True).index.names
```

```python
df.groupby(['model_year', 'cylinders']).mean(numeric_only=True)[['mpg', 'displacement']].head(12) # Có thể chọn col hiển thị
```

```python
# df.groupby('model_year').describe().transpose()
```

```python
year_cylinders = df.groupby(['model_year', 'cylinders']).mean(numeric_only=True)
year_cylinders.head(12)
```

```python
year_cylinders.index.levels
```

```python
year_cylinders.loc[70] # Lấy thông tin bên trong model_year = 70
```

```python
year_cylinders.loc[[70, 82]] # Lấy thông tin bên trong model_year = 70, 82
```

```python
year_cylinders.loc[(70, 4)] # Lấy được thông tin model_year = 70 và cylinders = 4
```

## GroupBy Operations - MultiIndex

```python
year_cylinders.head(12)
```

```python
year_cylinders.xs(key=70, level='model_year')
```

```python
# .xs() được sử dụng để chọn một nhóm con của DataFrame theo một chỉ mục cụ thể trong một cấp độ của MultiIndex
# key=4: Chọn các hàng có giá trị 4 ở cấp độ 'cylinders'
four_cylinders = year_cylinders.xs(key=4, level='cylinders') # Cách 1 (tối ưu khi làm việc với MultiIndex)
four_cylinders # Đặt tên biến thể hiện giá trị key và level (tips dễ đọc)
```

```python
year_cylinders.index.get_level_values('cylinders')
```

```python
year_cylinders.index.get_level_values('cylinders') == 4
```

```python
type(year_cylinders.index.get_level_values('cylinders') == 4)
```

```python
# Nếu sử dụng cách loc
year_cylinders.loc[year_cylinders.index.get_level_values('cylinders') == 4] # Cách 2 (Chậm hơn)
```

```python
# Nếu chỉ muốn lấy rows có cylinders = 6, 8 thì sao? Nên lọc trước khi thực hiện groupby 
df[df['cylinders'].isin([6, 8])].groupby(['model_year', 'cylinders']).mean(numeric_only=True)
```

```python
# agg sử dụng khi cần tổng hợp dữ liệu theo nhiều phép toán khác nhau
df_numeric = df.select_dtypes(include=['number']) # Chọn những cột dtypes = int, float
df_numeric.agg(['sum', 'mean', 'min', 'max']) # Áp dụng các hàm sum, mean, min, max lên toàn bộ các cột của df_numeric
```

```python
# Chỉ áp dụng các phép toán sum, mean, min, max trên cột 'cylinders' sau khi tính toán trên toàn bộ DataFrame.
# Kết quả trả về một Series chứa giá trị tổng hợp của 'cylinders'
df_numeric.agg(['sum', 'mean', 'min', 'max'])['cylinders']
```

```python
# Áp dụng các phép toán khác nhau cho từng cột
df_numeric.agg({'mpg' : ['max', 'mean'], 'weight': ['std', 'sum']})
```

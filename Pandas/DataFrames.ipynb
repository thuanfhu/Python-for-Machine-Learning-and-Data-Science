```python
myCols = ['total_bill', 'tip', 'day']
tipsDataFrame[myCols].head() # Phải có dấu [] nếu muốn truyền list cols
```

```python
(100 * tipsDataFrame['tip']) / tipsDataFrame['total_bill'] # Có thể thực hiện các phép toán giữa các cột
```

```python
tipsDataFrame['tips_percentage'] = (100 * tipsDataFrame['tip']) / tipsDataFrame['total_bill'] # Tạo một cột mới trong DataFrame
tipsDataFrame.head() # Nếu tên cột mới trùng với tên cột cũ thì cột cũ sẽ bị ghi đè giá trị
```

```python
tipsDataFrame['tips_percentage'] = np.round((100 * tipsDataFrame['tip']) / tipsDataFrame['total_bill']) # Làm tròn 2 chữ số thập phân với round()
tipsDataFrame.head()
```

```python
tipsDataFrame.drop('tips_percentage', axis=1).head() # Mặc định axis = 0 nghĩa là xóa hàng, axis = 1 là xóa cột
```

```python
tipsDataFrame.head() # Để ý mặc dù đã xóa tips_percentage rồi nhưng khi kiểm tra lại thì vẫn còn, chưa xóa vĩnh viễn?
```

```python
tipsDataFrame = tipsDataFrame.drop('tips_percentage', axis=1) # Xóa vĩnh viễn cột tips_percentage bằng cách gán lại DataFrame
tipsDataFrame.head() # Hoặc có thể sử dụng tham số inplace = True trong hàm drop
```

```python
tipsDataFrame.head()
```

```python
tipsDataFrame.index # RangeIndex là duy nhất, ML thường để mặc định
```

```python
tipsDataFrame = tipsDataFrame.set_index('Payment ID') # Set Payment ID thành Index, thay đổi là vĩnh viễn vì đã gán lại DataFrame
tipsDataFrame.head()
```

```python
tipsDataFrame = tipsDataFrame.reset_index() # Payment ID sẽ trở lại thành column, RangeIndex tăng dần theo mặc định 
tipsDataFrame.head()
```

```python
tipsDataFrame = tipsDataFrame.set_index('Payment ID')
tipsDataFrame.head()
```

```python
tipsDataFrame.iloc[0] # Lấy thông tin hàng đầu tiên theo RangeIndex
```

```python
tipsDataFrame.loc['Sun2959'] # Lấy thông tin hàng đầu tiên theo Payment ID
```

```python
tipsDataFrame.iloc[0:4] # Lấy thông tin theo RangeIndex hàng 1 đến 4 
```

```python
tipsDataFrame.loc[['Sun2959', 'Sun5260', 'Sun4608']] # Lấy thông tin nhiều hàng theo Payment ID
```

```python
tipsDataFrame.drop('Sun2959', axis=0) # Xóa hàng với ID = 'Sun2959', thay đổi chưa được lưu vĩnh viễn hoặc tipsDataFrame.iloc[1:] (slicing - tốt hơn)
```

```python
tipsDataFrame = tipsDataFrame.reset_index() # Chuyển Payment ID thành column, nếu không chuyển nó sẽ mất luôn Payment ID vì ignore_index=True
tipsDataFrame.tail()
```

```python
last_row = tipsDataFrame.iloc[-1] # Chọn hàng cuối cùng
last_row
```

```python
# last_row.to_frame().T: Chuyển Series thành một DataFrame với dạng hàng ngang (T là transpose)
# pd.concat([...], ignore_index=True): Ghép last_row vào tipsDataFrame và reset lại index
tipsDataFrame = pd.concat([tipsDataFrame, last_row.to_frame().T], ignore_index=True) # Chèn hàng cuối cùng vào table
```

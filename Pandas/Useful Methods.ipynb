```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('tips.csv')
df.head()
```

# Apply on Single Column

```python
str(123456788736)[-4:]
```

```python
def last_four(num):
    return int(str(num)[-4:])
last_four(12345567)
```

```python
# Khi cần áp dụng một hàm tùy chỉnh không có sẵn trong Pandas
# Khi cần thao tác trên cả dòng hoặc cả cột trong DataFrame
```

```python
df['last_four'] = df['CC Number'].apply(last_four)
df.head()
```

```python
# Khi hàm rất đơn giản, chỉ có một dòng (có thể có điều kiện) và không cần tái sử dụng hàm ở nhiều nơi -> Sử dụng lambda với apply
df['last_six'] = df['CC Number'].apply(lambda num: int(str(num)[-6:]))
df.head()
```

# Apply on Multiple Columns

```python
df.head()
```

```python
def quality(total_bill, tip):
    tip_percentage = (tip * 100) / total_bill
    if(tip_percentage >= 20):
        return "Excellent"
    elif(tip_percentage >= 12):
        return "Good"
    else:
        return "Not Good"
quality(100, 23)
```

```python
# Không cần phải định nghĩa riêng hàm riêng để gọi hàm quality nên dùng lambda
# Hàm apply sẽ tự động gọi hàm quality và chạy từng row để áp dụng tham số đầu vào hàm quality cho chính xác
df['Quality 1'] = df[['total_bill', 'tip']].apply(lambda df: quality(df['total_bill'], df['tip']), axis = 1) # Cách 1
df.head(10)
```

```python
df['Quality 2'] = np.vectorize(quality)(df['total_bill'], df['tip']) # Cách 2 (đơn giản và nhanh hơn)
df.head(10)
```

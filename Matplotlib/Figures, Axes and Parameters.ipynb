```python
import numpy as np
import matplotlib.pyplot as plt
```

## Prepare Data

```python
a = np.linspace(0, 10, 11)
a
```

```python
b = a ** 2
b
```

```python
x = np.arange(0, 10)
x
```

```python
y = x * 2
y
```

## Figures and Axes

```python
test_fig = plt.figure()
```

```python
test_fig_2 = plt.figure()

test_axes = test_fig_2.add_axes([0, 0, 0.5, 1]) # Tọa độ bắt đầu (0, 0) và w=0.5, h=1 (tính theo figure)
```

```python
fig = plt.figure()

# LARGE AXES
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)

axes1.set_title('Power of 4')
axes1.set_xlabel('A axis')
axes1.set_ylabel('B axis')
axes1.set_xlim(0, 10)
axes1.set_ylim(0, 100)

# SMALL AXES
axes2 = fig.add_axes([0.2, 0.4, 0.3, 0.3])
axes2.plot(a, b)

axes2.set_title('Zoomed in')
axes2.set_xlabel('A axis')
axes2.set_ylabel('B axis')
axes2.set_xlim(4, 10)
axes2.set_ylim(0, 80)

plt.show()
```

## Figure Parameters

```python
# figsize=(5, 2): Kích thước hình vẽ (5 inch rộng, 2 inch cao)
# dpi=150: Độ phân giải 150 điểm ảnh trên mỗi inch (càng cao thì hình ảnh càng chất lượng nhưng tốn bộ nhớ và RAM)
# fig.savefig('my_figure.jpeg'): Giữ nguyên toàn bộ vùng vẽ (có thể dư khoảng trắng)
# fig.savefig('my_figure_tight.jpeg', bbox_inches='tight'): Cắt bớt khoảng trắng không cần thiết xung quanh biểu đồ (Tự động cắt khoảng trắng dư)
```

```python
fig2 = plt.figure(figsize=(5, 2), dpi=150)

axes3 = fig2.add_axes([0, 0, 1, 1])
axes3.plot(a, b);

fig2.savefig('my_figure.jpeg', bbox_inches='tight')
```

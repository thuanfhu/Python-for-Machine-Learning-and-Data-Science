```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 10)
x
```

```python
y = x * 2
y
```

```python
plt.plot(x, y) # Để ý ở đây sẽ trả về danh sách các đối tượng Line2D trong bộ nhớ và vẽ đồ thị
```

```python
plt.plot(x, y)
plt.show() # Nếu sử dụng file .py thì thêm dòng này. Trong notebook không trả về danh sách các đối tượng Line2D trong bộ nhớ, chỉ vẽ đồ thị
```

```python
plt.plot(x, y)  # Vẽ đồ thị với x trên trục hoành và y trên trục tung

plt.title('Tên của biểu đồ')  # Đặt tiêu đề cho biểu đồ
plt.xlabel('X Axis')  # Ghi nhãn trục X
plt.ylabel('Y Axis')  # Ghi nhãn trục Y

plt.xlim(0, 6)  # Giới hạn trục X từ 0 đến 6
plt.ylim(0, 15)  # Giới hạn trục Y từ 0 đến 15

plt.savefig('myfirstplot.png') # Nếu chỉ ghi tên file, nó sẽ lưu chung nơi viết notebook này (còn không thì truyền path cụ thể)
```

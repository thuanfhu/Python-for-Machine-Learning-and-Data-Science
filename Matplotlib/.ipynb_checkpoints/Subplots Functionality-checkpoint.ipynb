```python
import numpy as np
import matplotlib.pyplot as plt
```

## Prepare Data

```python
a = np.linspace(0, 10, 11)
b = a ** 4
```

```python
x = np.arange(0, 10)
y = x * 2
```

## Subplots

```python
# subplot được dùng để tạo nhiều biểu đồ (axes) trong cùng một figure
fig1, axes1 = plt.subplots()
```

```python
fig3, axes3 = plt.subplots(nrows=2, ncols=4)

axes3[0][0].plot(x, y)
axes3[0][1].plot(a, b)
axes3[-1][-1].plot(x, y)
axes3[-1][-2].plot(a, b)

# Tự động căn chỉnh khoảng cách giữa các subplot để không bị chồng lấn.  
fig3.tight_layout()
```

```python
# Nếu kiểm tra axes4 sẽ là 1D array nên có thể dùng for loop
fig4, axes4 = plt.subplots(nrows=3, ncols=1)
for ax in axes4:
    ax.plot(x, y)

fig4.tight_layout()
```

```python
axes4
```

```python
# Có thể set width và height thêm bằng cách: fig5.set_figwidth(10) và fig5.set_figheight(5) hoặc dùng figsize
fig5, axes5 = plt.subplots(nrows=2, ncols=2, figsize=(10, 5), dpi=150)

for row in axes5:
    for col in row:
        col.plot(x, y)

axes5[-1][-1].plot(a, b)
fig5.tight_layout() # Muốn căn chỉnh thêm thì sao?

# wspace và hspace giúp điều chỉnh khoảng cách giữa các subplot
# left, right, bottom, top giúp căn chỉnh lề của figure
fig5.subplots_adjust(wspace=0.75, hspace=0.5, left=0.1, right=0.9, bottom=0.1, top=0.9) # Tất cả giá trị nên nằm trong [0,1] để giữ figure cân đối

# Đặt tiêu đề chung
fig5.suptitle('Four subplots', fontsize=18)

# Đặt tiêu đề, tên trục riêng cho subplot
axes5[0][0].set_title('Subplot 1', fontsize=8)
axes5[0][0].set_xlim(0, 20)
axes5[0][0].set_ylim(0, 10)
axes5[0][0].set_xlabel('X Axis - Subplot 1')
axes5[0][0].set_ylabel('Y Axis - Subplot 1')

fig5.savefig('my_subplots.jpeg', bbox_inches='tight')
```

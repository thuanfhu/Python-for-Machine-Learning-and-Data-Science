```python
import numpy as np
import matplotlib.pyplot as plt
```

## Prepare Data

```python
x = np.linspace(0, 10, 11)
x
```

## Legends

```python
fig1 = plt.figure()

axes1 = plt.axes([0, 0, 1, 1])

axes1.plot(x, x, label='X vs X')
axes1.plot(x, x**2, label='X vs X^2')
axes1.plot(x, 2*x, label='X vs 2X')

axes1.legend(loc='center')
```

```python
fig2 = plt.figure()

axes2 = plt.axes([0, 0, 1, 1])

axes2.plot(x, x, label='X vs X')
axes2.plot(x, x**2, label='X vs X^2')
axes2.plot(x, 2*x, label='X vs 2X')

# loc=(x, y): Xc nh v tr ca legend theo ta  figure
axes2.legend(loc=(1.1, 0.5)) # Legend s nm bn phi biu , cn gia theo trc y
```

## Colors and Styles

```python
fig3 = plt.figure()

axes3 = plt.axes([0, 0, 1, 1])

# color parameter dng  thay i mu sc axes
# lw hoc linewidth parameter dng  chnh sa  dy axes (mc nh l 1)
axes3.plot(x, x, color='red', label='X vs X', lw=4)
axes3.plot(x, x+10, color='#0ff007', label='X vs X+10', linewidth=0.5)
axes3.plot(x, x**2, color='#b211f7', label='X vs X^2')

# possible ls or linestyle options: --, -, -., :, steps
axes3.plot(x+1, x**2, color='pink', label='X+1 vs X^2', lw=3, ls='-.')
axes3.plot(x+2, x**2, color='blue', label='X+2 vs X^2', lw=3, linestyle=':')

# custom linestyle
customls = axes3.plot(x, x+20, color='orange', label='X vs X+20', lw=3)

# set_dashes() cho php thit lp mt style gch t ty chnh bng cch truyn mt danh sch s vo.
# cu trc ca set_dashes([dash_length, gap_length, dash_length, gap_length, ...])
customls[0].set_dashes([1, 2, 1, 2, 10, 2]) # 1 n v di, khong trng, 1 v di, khong trng, 10 v di, khong trng,... (lp li)

axes3.legend()
```

```python
fig4 = plt.figure()

axes4 = plt.axes([0, 0, 1, 1])

# marker  hin th cc im d liu trn ng  th v ms (markersize)  iu chnh kch thc ca chng
axes4.plot(x, x, color='red', label='X vs X', lw=2, ls='--', marker='o', markersize=10)
axes4.plot(x, x+2, color='blue', label='X vs X+2', ls='--', marker='*', ms=20)
axes4.plot(x, x-2, color='#00ff11', label='X vs X-2', ls=':', marker='P', ms=15)

# markerfacecolor='color': i mu bn trong marker (fill color)
axes4.plot(x, x-4, color='#00ff11', label='X vs X-4', ls=':', marker='P', ms=15, markerfacecolor='orange')

# markeredgecolor='color': i mu vin ngoi marker
axes4.plot(x, x-6, color='#00ff11', label='X vs X-6', ls=':', marker='P', ms=15, markerfacecolor='orange', markeredgecolor='red')

# markeredgewidth=n: iu chnh  dy vin (s pixel)
axes4.plot(x, x-8, color='#00ff11', label='X vs X-8', ls=':', marker='P', ms=15, markerfacecolor='orange', markeredgecolor='red', markeredgewidth=2)

axes4.legend(handleheight=2, loc=(1.05, 0.25)) # handleheight parameter tng khong cch gia cc mc trong legend
```
